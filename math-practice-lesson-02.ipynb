Задаются уравнением $\alpha'' + w^2 \alpha = 0$, где $\alpha$ - угол отклонения. 


![](http://ru.solverbook.com/my_images/pic1103.png)



```python
import matplotlib.pyplot as plt
import numpy as np

# Константы, начальные параметры
gravit_acc = 9.8
length = 1
alpha0 = 0.01


# Уравнение для alpha(t):=
def alpha(t):
    return alpha0*np.cos(np.sqrt(gravit_acc/length)*t)



# Рисуем график
t = np.linspace(0, 3)

plt.plot(t, alpha(t), color = 'green')
plt.title('Грфик alpha(t)')

plt.grid(True)
plt.show()
```
